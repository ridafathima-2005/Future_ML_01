```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
df = pd.read_csv("Store sales.csv", encoding='latin1')
```

```python
df = df.dropna()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
df = df.sort_values('Order Date')
```

```python
daily_sales = df.groupby('Order Date')['Sales'].sum().reset_index()
```

```python
daily_sales.shape
```

```python
plt.figure(figsize=(10,5))
plt.plot(daily_sales['Order Date'], daily_sales['Sales'])
plt.title("Sales Over Time")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.show()
```

```python
daily_sales['Days'] = (daily_sales['Order Date'] - daily_sales['Order Date'].min()).dt.days
```

```python
X = daily_sales[['Days']]
y = daily_sales['Sales']
```

```python
print(X.shape)
print(y.shape)
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
y_pred = model.predict(X)
```

```python
mae = mean_absolute_error(y, y_pred)
rmse = np.sqrt(mean_squared_error(y, y_pred))

print("MAE:", mae)
print("RMSE:", rmse)
```

```python
plt.figure(figsize=(10,5))
plt.plot(daily_sales['Order Date'], y, label='Actual')
plt.plot(daily_sales['Order Date'], y_pred, label='Predicted')
plt.legend()
plt.title("Actual vs Predicted Sales")
plt.show()
```

```python
future_days = np.arange(X['Days'].max()+1, X['Days'].max()+30).reshape(-1,1)
future_sales = model.predict(future_days)
```

```python
plt.figure(figsize=(10,5))
plt.plot(future_sales)
plt.title("Future Sales Prediction (Next 30 Days)")
plt.xlabel("Days Ahead")
plt.ylabel("Predicted Sales")
plt.show()
```
